# **Завдання 2**

```python
import networkx as nx

# Створення графа
G = nx.Graph()

# Додавання вершин (станцій метро)
stations = ["Akademmistechko", "Zhytomyrska", "Sviatoshyn", "Nyvky", "Beresteiska",
            "Shuliavska", "Politekhnichnyi Instytut", "Vokzalna", "Universytet", "Teatralna",
            "Khreshchatyk", "Arsenalna", "Dnipro", "Hidropark", "Livoberezhna",
            "Darnytsia", "Chernihivska", "Lisova", "Heroiv Dnipra", "Minska",
            "Obolon", "Pochaina", "Tarasa Shevchenka", "Kontraktova Ploshcha", "Poshtova Ploshcha",
            "Maidan Nezalezhnosti", "Ploshcha Lva Tolstoho", "Olimpiiska", "Palats Ukraina", "Lybidska",
            "Demiivska", "Holosiivska", "Vasylkivska", "Vystavkovyi Tsentr", "Ipodrom",
            "Teremky"]

G.add_nodes_from(stations)

# Додавання ребер (ліній метро)
edges = [("Akademmistechko", "Zhytomyrska"), ("Zhytomyrska", "Sviatoshyn"), ("Sviatoshyn", "Nyvky"),
         ("Nyvky", "Beresteiska"), ("Beresteiska", "Shuliavska"), ("Shuliavska", "Politekhnichnyi Instytut"),
         ("Politekhnichnyi Instytut", "Vokzalna"), ("Vokzalna", "Universytet"), ("Universytet", "Teatralna"),
         ("Teatralna", "Khreshchatyk"), ("Khreshchatyk", "Arsenalna"), ("Arsenalna", "Dnipro"),
         ("Dnipro", "Hidropark"), ("Hidropark", "Livoberezhna"), ("Livoberezhna", "Darnytsia"),
         ("Darnytsia", "Chernihivska"), ("Chernihivska", "Lisova"),
         ("Heroiv Dnipra", "Minska"), ("Minska", "Obolon"), ("Obolon", "Pochaina"),
         ("Pochaina", "Tarasa Shevchenka"), ("Tarasa Shevchenka", "Kontraktova Ploshcha"),
         ("Kontraktova Ploshcha", "Poshtova Ploshcha"), ("Poshtova Ploshcha", "Maidan Nezalezhnosti"),
         ("Maidan Nezalezhnosti", "Ploshcha Lva Tolstoho"), ("Ploshcha Lva Tolstoho", "Olimpiiska"),
         ("Olimpiiska", "Palats Ukraina"), ("Palats Ukraina", "Lybidska"), ("Lybidska", "Demiivska"),
         ("Demiivska", "Holosiivska"), ("Holosiivska", "Vasylkivska"), ("Vasylkivska", "Vystavkovyi Tsentr"),
         ("Vystavkovyi Tsentr", "Ipodrom"), ("Ipodrom", "Teremky")]

G.add_edges_from(edges)

# Алгоритм DFS
def dfs_paths(graph, start, goal, path=None):
    if path is None:
        path = [start]
    if start == goal:
        yield path
    for next_vertex in set(graph[start]) - set(path):
        yield from dfs_paths(graph, next_vertex, goal, path + [next_vertex])

# Алгоритм BFS
def bfs_paths(graph, start, goal):
    queue = [(start, [start])]
    while queue:
        (vertex, path) = queue.pop(0)
        for next_vertex in set(graph[vertex]) - set(path):
            if next_vertex == goal:
                yield path + [next_vertex]
            else:
                queue.append((next_vertex, path + [next_vertex]))

# Знаходження шляхів для станцій "Akademmistechko" та "Teremky" з використанням DFS та BFS
dfs_result = list(dfs_paths(G, "Akademmistechko", "Vokzalna"))
bfs_result = list(bfs_paths(G, "Akademmistechko", "Vokzalna"))

print("DFS Paths:")
for path in dfs_result:
    print(path)

print("\nBFS Paths:")
for path in bfs_result:
    print(path)
```

У вищенаведених шляхах, отриманих за допомогою алгоритмів DFS і BFS, ми бачимо, що обидва алгоритми вказують на однаковий шлях:

DFS Paths: ['Akademmistechko', 'Zhytomyrska', 'Sviatoshyn', 'Nyvky', 'Beresteiska', 'Shuliavska', 'Politekhnichnyi Instytut', 'Vokzalna']
BFS Paths: ['Akademmistechko', 'Zhytomyrska', 'Sviatoshyn', 'Nyvky', 'Beresteiska', 'Shuliavska', 'Politekhnichnyi Instytut', 'Vokzalna']
Отже, обидва алгоритми знайшли однаковий шлях від станції "Akademmistechko" до станції "Vokzalna".

Різниця між алгоритмами полягає у порядку обходу вершин:

DFS (Depth-First Search) обходить граф "вглиб", спочатку досліджуючи одну гілку в глибину, перед тим як переходити до інших гілок. Таким чином, шляхи, знайдені за допомогою DFS, можуть бути не найкоротшими, але цей алгоритм може швидше знаходити розв'язок, оскільки перевіряє близькі вершини першими.

BFS (Breadth-First Search) обходить граф "вшир", розглядаючи всі сусідні вершини перед тим, як перейти до наступного рівня. Тому BFS гарантує знаходження найкоротшого шляху, але може вимагати більше часу на обробку, особливо великих графів.

У данному випадку обидва алгоритми знаходять один і той же шлях, оскільки граф представляє собою просту послідовність станцій без додаткових розгалужень на одній гілці метрополітену.

```python
import networkx as nx

# Заданий граф
G = nx.Graph()

# Додані вершини та ребра (з попереднього завдання)
stations = ["Akademmistechko", "Zhytomyrska", "Sviatoshyn", "Nyvky", "Beresteiska",
            "Shuliavska", "Politekhnichnyi Instytut", "Vokzalna", "Universytet", "Teatralna",
            "Khreshchatyk", "Arsenalna", "Dnipro", "Hidropark", "Livoberezhna",
            "Darnytsia", "Chernihivska", "Lisova", "Heroiv Dnipra", "Minska",
            "Obolon", "Pochaina", "Tarasa Shevchenka", "Kontraktova Ploshcha", "Poshtova Ploshcha",
            "Maidan Nezalezhnosti", "Ploshcha Lva Tolstoho", "Olimpiiska", "Palats Ukraina", "Lybidska",
            "Demiivska", "Holosiivska", "Vasylkivska", "Vystavkovyi Tsentr", "Ipodrom",
            "Teremky"]

edges = [("Akademmistechko", "Zhytomyrska", {'weight': 3}), ("Zhytomyrska", "Sviatoshyn", {'weight': 2}),
         ("Sviatoshyn", "Nyvky", {'weight': 4}), ("Nyvky", "Beresteiska", {'weight': 5}),
         ("Beresteiska", "Shuliavska", {'weight': 3}), ("Shuliavska", "Politekhnichnyi Instytut", {'weight': 4}),
         ("Politekhnichnyi Instytut", "Vokzalna", {'weight': 3}), ("Vokzalna", "Universytet", {'weight': 2}),
         ("Universytet", "Teatralna", {'weight': 1}), ("Teatralna", "Khreshchatyk", {'weight': 2}),
         ("Khreshchatyk", "Arsenalna", {'weight': 3}), ("Arsenalna", "Dnipro", {'weight': 2}),
         ("Dnipro", "Hidropark", {'weight': 3}), ("Hidropark", "Livoberezhna", {'weight': 4}),
         ("Livoberezhna", "Darnytsia", {'weight': 5}), ("Darnytsia", "Chernihivska", {'weight': 3}),
         ("Chernihivska", "Lisova", {'weight': 4}), ("Heroiv Dnipra", "Minska", {'weight': 3}),
         ("Minska", "Obolon", {'weight': 2}), ("Obolon", "Pochaina", {'weight': 3}),
         ("Pochaina", "Tarasa Shevchenka", {'weight': 4}), ("Tarasa Shevchenka", "Kontraktova Ploshcha", {'weight': 2}),
         ("Kontraktova Ploshcha", "Poshtova Ploshcha", {'weight': 3}), ("Poshtova Ploshcha", "Maidan Nezalezhnosti", {'weight': 2}),
         ("Maidan Nezalezhnosti", "Ploshcha Lva Tolstoho", {'weight': 3}), ("Ploshcha Lva Tolstoho", "Olimpiiska", {'weight': 4}),
         ("Olimpiiska", "Palats Ukraina", {'weight': 2}), ("Palats Ukraina", "Lybidska", {'weight': 3}),
         ("Lybidska", "Demiivska", {'weight': 2}), ("Demiivska", "Holosiivska", {'weight': 3}),
         ("Holosiivska", "Vasylkivska", {'weight': 4}), ("Vasylkivska", "Vystavkovyi Tsentr", {'weight': 2}),
         ("Vystavkovyi Tsentr", "Ipodrom", {'weight': 3}), ("Ipodrom", "Teremky", {'weight': 4})]

G.add_nodes_from(stations)
G.add_edges_from(edges)

# Алгоритм Дейкстри
def dijkstra(graph, start):
    # Ініціалізація відстаней до всіх вершин як нескінченні
    distance = {node: float('inf') for node in graph.nodes()}
    # Відстань від початкової вершини до себе дорівнює 0
    distance[start] = 0
    # Відвідані вершини
    visited = set()
    # Попередники для відновлення шляху
    predecessors = {}

    # Поки є неперевірені вершини
    while len(visited) < len(graph.nodes()):
        # Знайдемо вершину з найменшою відстанню серед неперевірених
        current_node = min((node for node in graph.nodes() if node not in visited), key=lambda x: distance[x])
        visited.add(current_node)

        # Оновимо відстані до сусідів поточної вершини
        for neighbor in graph.neighbors(current_node):
            if neighbor not in visited:
                new_distance = distance[current_node] + graph[current_node][neighbor]['weight']
                if new_distance < distance[neighbor]:
                    distance[neighbor] = new_distance
                    predecessors[neighbor] = current_node

    return distance, predecessors

from itertools import combinations

# Знаходження найкоротших шляхів між всіма парами вершин
all_shortest_paths = {}
for start_node in G.nodes():
    distances, predecessors = dijkstra(G, start_node)
    all_shortest_paths[start_node] = distances

# Виведення результатів
for start_node, distances in all_shortest_paths.items():
    print(f"Shortest paths from {start_node}:")
    for end_node, distance in distances.items():
        print(f"To {end_node}: {distance}")
```
